### a função \_\_init\_\_():
---

```python
class Pessoa:
    def __init__(self,nome,sobrenome,idade):                      # o parametro 'self' pode ser trocado por qualquer outro.
        self.nome = nome
        self.sobrenome = sobrenome
        self.idade = idade
        
    def aniversario(self):
        self.idade = self.idade + 1

pessoa1 = Pessoa("fulano",26)

print(pessoa1.nome)
print(pessoa1.idade)
pessoa1.aniversario()
print(pessoa1.idade)
```

### adicionando o função __init__():

```python
class Estudante2(Pessoa):
    def __init__(self, nome, sobrenome,idade):                   # Estudando não herda a função de inicialização de Pessoa
        Pessoa.__init__(self, nome, sobrenome, idade)             # Herda a função __init__ de Pessoa
        # super.__init__(self,nome, sobrenome,idade)             # herda todas as propriedades e métodos da classe pai

o2 = Estudante2("joão","pereira", 28)
print(o2.sobrenome)
```

### adicionando uma propriedade e um método:
---

```python
class conta:
    def __init__(self, num, sal):
        self.numero = num
        self.saldo = sal
        
    def imprimeSal(self):
        print(self.saldo)
        
class poupanca(conta):
    def __init__(self, num, sal, ani):
        super().__init__(num,sal)
        self.aniversario = ani                            # nova propriedade
        
    def imprimeAni(self):                                 # novo método
        print(self.aniversario)
        
o5 = poupanca(2000,100,"1-2-2020")
o5.imprimeSal()
o5.imprimeAni()
```
